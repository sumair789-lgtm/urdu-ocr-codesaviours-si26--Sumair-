```python
# Step 1 — Connect Google Drive (permanent storage)
from google.colab import drive
drive.mount('/content/drive')

# Step 2 — Make a folder inside Drive (won't disappear on session reset)
import os
save_dir = '/content/drive/MyDrive/urdu-ocr-si26/data/synthetic'
os.makedirs(save_dir, exist_ok=True)

# Step 3 — Re-download font (since Colab storage was wiped)
import urllib.request
font_path = 'urdu_font.ttf'
if not os.path.exists(font_path):
    font_url = "https://github.com/googlefonts/noto-fonts/raw/main/hinted/ttf/NotoNastaliqUrdu/NotoNastaliqUrdu-Regular.ttf"
    urllib.request.urlretrieve(font_url, font_path)
print("Font ready:", os.path.getsize(font_path), "bytes")

# Step 4 — Generate 20 images, saved directly to Drive
from PIL import Image, ImageDraw, ImageFont
from IPython.display import display

urdu_texts = [
    "پاکستان زندہ باد", "آج کا موسم خوشگوار ہے", "تعلیم ہر انسان کا حق ہے",
    "کراچی پاکستان کا سب سے بڑا شہر ہے", "محنت کامیابی کی کنجی ہے",
    "صحت سب سے بڑی نعمت ہے", "وقت کی قدر کرنی چاہیے", "لاہور پاکستان کا تاریخی شہر ہے",
    "کتاب علم کا خزانہ ہے", "والدین کی عزت کرنی چاہیے", "اردو ہماری قومی زبان ہے",
    "پاکستان کا دارالحکومت اسلام آباد ہے", "محنتی انسان کبھی ناکام نہیں ہوتا",
    "دوستی ایک قیمتی رشتہ ہے", "صبر کا پھل میٹھا ہوتا ہے", "علم بغیر عمل کے بیکار ہے",
    "پاکستان کا قومی پھل آم ہے", "ایمانداری بہترین پالیسی ہے",
    "ورزش صحت کے لیے ضروری ہے", "ہر انسان کو محنت کرنی چاہیے"
]

font = ImageFont.truetype(font_path, 36, layout_engine=ImageFont.Layout.RAQM)
generated = []

for i, text in enumerate(urdu_texts):
    img = Image.new('RGB', (500, 90), color='white')
    draw = ImageDraw.Draw(img)
    draw.text((480, 20), text, fill='black', font=font, direction='rtl', language='ur', anchor='ra')
    path = f'{save_dir}/urdu_{i+1}.png'
    img.save(path)
    generated.append(path)
    print(f'Saved: urdu_{i+1}.png -- {text}')

print(f'\nDone! {len(generated)} images saved permanently to Google Drive.')
display(Image.open(generated[0]))
```

What is OCR?
Optical Character Recognition is besically a technology in wich the text from images and scanned documents are converted to editable and machine readable text.

Why Urdu OCR is harder then the English OCR?
As the Urdu Language Script is more cursive then English and the words are joined together with no proper spacing so it becomes difficult to recognize those letter by the computer.

2 Real World Use Cases
1: Preservation of old Pakistani News Papers like Jung etc
2: For the Visually impaired people that helps them to converte text-to-speech for becoming an accessibility tools.

```python
import csv, os

# Purani CSV delete karo
if os.path.exists('data/labels.csv'):
    os.remove('data/labels.csv')
    print("Old CSV deleted!")

labels = []

synthetic_texts = [
    "پاکستان زندہ باد", "آج کا موسم خوشگوار ہے", "تعلیم ہر انسان کا حق ہے",
    "کراچی پاکستان کا سب سے بڑا شہر ہے", "محنت کامیابی کی کنجی ہے",
    "صحت سب سے بڑی نعمت ہے", "وقت کی قدر کرنی چاہیے", "لاہور پاکستان کا تاریخی شہر ہے",
    "کتاب علم کا خزانہ ہے", "والدین کی عزت کرنی چاہیے", "اردو ہماری قومی زبان ہے",
    "پاکستان کا دارالحکومت اسلام آباد ہے", "محنتی انسان کبھی ناکام نہیں ہوتا",
    "دوستی ایک قیمتی رشتہ ہے", "صبر کا پھل میٹھا ہوتا ہے", "علم بغیر عمل کے بیکار ہے",
    "پاکستان کا قومی پھل آم ہے", "ایمانداری بہترین پالیسی ہے",
    "ورزش صحت کے لیے ضروری ہے", "ہر انسان کو محنت کرنی چاہیے"
]

for i, text in enumerate(synthetic_texts):
    labels.append({'image': f'data/raw/synthetic/urdu_{i+1}.png', 'text': text})

for folder in ['newspaper', 'books', 'signboards', 'other']:
    folder_path = f'data/raw/{folder}'
    if os.path.exists(folder_path):
        for img in os.listdir(folder_path):
            labels.append({'image': f'data/raw/{folder}/{img}', 'text': 'TODO'})

with open('data/labels.csv', 'w', newline='', encoding='utf-8') as f:
    writer = csv.DictWriter(f, fieldnames=['image', 'text'])
    writer.writeheader()
    writer.writerows(labels)

# Drive mein bhi save karo
import shutil
shutil.copy('data/labels.csv', '/content/drive/MyDrive/urdu-ocr-si26/data/labels.csv')

print(f'New CSV created with {len(labels)} entries!')
```
